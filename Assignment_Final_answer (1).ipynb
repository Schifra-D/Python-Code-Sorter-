```python
import pandas as pd
```

```python
column_names = ['Employee ID', 'Sample', 'Strength', 'Weakness']
```

```python
df = pd.read_csv('/content/practice.tsv', sep='\t',header=None, names=column_names)
```

```python
df.head()
```

```python
##FINAL CODE

# List of words to remove from the 'Strength' and 'Weakness' columns
items_to_remove = ['lies', 'in', 'is', 'the', 'by', 'i', 'my']

# Replace ':' and '-' with ','
df['Sample'] = df['Sample'].str.replace(r'[:\-]', ',')

# Replace ' and ' with ', '
df['Sample'] = df['Sample'].str.replace(' and ', ', ')

# Convert 'Strengths' to 'Strength'
df['Sample'] = df['Sample'].str.replace('Strengths', 'Strength')

# Loop through the 'Sample' column to extract strength and weakness
for index, row in df.iterrows():
    sample = row['Sample'].lower()  # Convert to lowercase for case insensitivity
    strength_idx = sample.find('strength')
    weakness_idx = sample.find('weakness')

    if strength_idx != -1 and weakness_idx != -1:
        # If both "Strength" and "Weakness" are found
        if strength_idx < weakness_idx:
            # "Strength" occurs before "Weakness"
            strength_text = sample[strength_idx + len('strength'):weakness_idx].strip()
            weakness_text = sample[weakness_idx + len('weakness'):].strip()
        else:
            # "Weakness" occurs before "Strength"
            weakness_text = sample[weakness_idx + len('weakness'):strength_idx].strip()
            strength_text = sample[strength_idx + len('strength'):].strip()
    elif strength_idx != -1:
        # Only "Strength" is found
        strength_text = sample[strength_idx + len('strength'):].strip()
        weakness_text = None
    elif weakness_idx != -1:
        # Only "Weakness" is found
        weakness_text = sample[weakness_idx + len('weakness'):].strip()
        strength_text = None
    else:
        # Neither "Strength" nor "Weakness" is found
        strength_text = None
        weakness_text = None

    # Remove specified words from 'Strength' and 'Weakness' using items_to_remove
    if strength_text:
        strength_text = ' '.join(word if word.startswith(',') else word for word in strength_text.split() if word not in items_to_remove)
    if weakness_text:
        weakness_text = ' '.join(word if word.startswith(',') else word for word in weakness_text.split() if word not in items_to_remove)

    # Add space after a comma if there is none
    if strength_text:
        strength_text = strength_text.replace(',', ', ') if ',' in strength_text else strength_text
    if weakness_text:
        weakness_text = weakness_text.replace(',', ', ') if ',' in weakness_text else weakness_text

    # Remove the comma (',') at the beginning and end of a line if it starts or ends with a comma
    if strength_text and strength_text.startswith(', '):
        strength_text = strength_text[2:]
    if weakness_text and weakness_text.startswith(', '):
        weakness_text = weakness_text[2:]
    if strength_text and strength_text.endswith(', '):
        strength_text = strength_text[:-2]
    if weakness_text and weakness_text.endswith(', '):
        weakness_text = weakness_text[:-2]

    # Replace double whitespace with a single whitespace
    if strength_text:
        strength_text = ' '.join(strength_text.split())
    if weakness_text:
        weakness_text = ' '.join(weakness_text.split())

    # Convert the first letter of every word to uppercase
    if strength_text:
        strength_text = strength_text.title()
    if weakness_text:
        weakness_text = weakness_text.title()

    # Update the 'Strength' and 'Weakness' columns
    df.at[index, 'Strength'] = strength_text
    df.at[index, 'Weakness'] = weakness_text

# Set the option to display full column content without truncation
pd.set_option('display.max_colwidth', None)

# Print the updated DataFrame with 'Employee ID' as the index
print(df.set_index('Employee ID')[['Sample', 'Strength', 'Weakness']])

```
